```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df.columns = df.columns.str.strip()
df.drop(columns =['name', 'Unnamed: 0'])
df = df.drop(columns =['name', 'Unnamed: 0'])
df
```


The non-predictive columns (e.g., player names, IDs) was dropped so that noise or data leakage is completely removed.

```python
X= df.drop(columns=['target_5yrs'])
y = df['target_5yrs']
```

target_5yrs column is clearly defined as the dependent variable while others are the independent variable.

```python
corr_matrix =df.corr()
target_corr = corr_matrix ['target_5yrs'].sort_values(ascending= False)
print(target_corr)

```

Correlation analysis was performed to identify highly correlated features and reduce redundancy. The three highly correlated factors are  gp, min ,fgm

```python
df_highly_correlated = df[['target_5yrs' , 'gp', 'min','fgm',]]
highly_correlated = ['target_5yrs' , 'gp', 'min','fgm',]
corr_matrix = df_highly_correlated.corr()
plt.figure(figsize=(18, 8))
sns.heatmap( corr_matrix ,annot =True,cmap = 'coolwarm', xticklabels= highly_correlated ,
    yticklabels= highly_correlated)
plt.show()
```
